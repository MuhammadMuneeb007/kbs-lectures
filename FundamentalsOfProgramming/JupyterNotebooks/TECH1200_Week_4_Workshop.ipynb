````python
def get_public_methods(obj):
    return [method for method in dir(obj) if callable(getattr(obj, method)) and not method.startswith('__')]

import numpy as np
import pandas as pd


# For Integer (`int`)
a = 5
int_methods = get_public_methods(a)
print("Integer methods:", int_methods)

# For String (`str`)
a = "example"
str_methods = get_public_methods(a)
print("String methods:", str_methods)

# For List (`list`)
a = [1, 2, 3]
list_methods = get_public_methods(a)
print("List methods:", list_methods)

# For Tuple (`tuple`)
a = (1, 2, 3)
tuple_methods = get_public_methods(a)
print("Tuple methods:", tuple_methods)

# For Dictionary (`dict`)
a = {'key1': 'value1', 'key2': 'value2'}
dict_methods = get_public_methods(a)
print("Dictionary methods:", dict_methods)

# For Set (`set`)
a = {1, 2, 3}
set_methods = get_public_methods(a)
print("Set methods:", set_methods)

# For Frozenset (`frozenset`)
a = frozenset([1, 2, 3])
frozenset_methods = get_public_methods(a)
print("Frozenset methods:", frozenset_methods)

# For Bytes (`bytes`)
a = b"example"
bytes_methods = get_public_methods(a)
print("Bytes methods:", bytes_methods)

# For Bytearray (`bytearray`)
a = bytearray(b"example")
bytearray_methods = get_public_methods(a)
print("Bytearray methods:", bytearray_methods)
````
